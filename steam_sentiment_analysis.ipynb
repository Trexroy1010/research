```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

def get_preprocessed_data():
    """
    Loads the DataFrame from a CSV file, preprocesses it, and returns the
    training and testing sets.

    You MUST replace 'your_data_file.csv' with the actual path to your file.
    For example: 'C:/Users/YourName/Desktop/your_data_file.csv'
    """
    try:
        # --- IMPORTANT: REPLACE 'your_data_file.csv' WITH YOUR FILE PATH ---
        df = merged_game_data

    except FileNotFoundError:
        print("Error: The file was not found. Please check the file path.")
        return None, None, None, None
    except Exception as e:
        print(f"An error occurred while loading the file: {e}")
        return None, None, None, None

    # Identify all numeric columns (int64 and float64) from the DataFrame info you provided
    numeric_cols = [
        'appid', 'votes_up', 'votes_funny', 'timestamp_created', 'author_playtime_forever',
        'price_x', 'word_count', 'numeric_sentiment', 'log_playtime', 'interaction',
        'positive', 'negative', 'userscore', 'average_forever', 'average_2weeks',
        'median_forever', 'median_2weeks', 'price_y', 'initialprice', 'discount', 'ccu'
    ]

    # Check if all numeric columns are present in the DataFrame
    if not all(col in df.columns for col in numeric_cols):
        print("Error: The DataFrame is missing one or more of the expected numeric columns.")
        return None, None, None, None

    # Define features (X) and target (y)
    # X will be all numeric columns except the target 'numeric_sentiment'
    X = df[numeric_cols].drop('numeric_sentiment', axis=1)
    y = df['numeric_sentiment']

    # --- IMPORTANT FIX: Re-map sentiment labels to be a contiguous range from 0 ---
    # This resolves the `IndexError` from `to_categorical`
    y, _ = pd.factorize(y, sort=True)
    # ---------------------------------------------------------------------------

    # Handle missing values by imputing the mean
    imputer = SimpleImputer(strategy='mean')
    X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X_imputed, y, test_size=0.2, random_state=42, stratify=y
    )

    # Print the shape and unique values to confirm the fix
    print(f"X_train shape: {X_train.shape}")
    print(f"X_test shape: {X_test.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"y_test shape: {y_test.shape}")
    print(f"Unique values in y_train: {np.unique(y_train)}")

    return X_train, X_test, y_train, y_test

if __name__ == '__main__':
    X_train, X_test, y_train, y_test = get_preprocessed_data()
    if X_train is not None:
        print("\nData preprocessing complete and datasets are ready for modeling.")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
#from data_preprocessing import get_preprocessed_data
from tensorflow.keras.utils import to_categorical

# Load the preprocessed data
X_train, X_test, y_train, y_test = get_preprocessed_data()

# IMPORTANT: Determine the number of unique sentiment classes.
# This assumes your 'numeric_sentiment' column has integer labels from 0 to N-1.
num_classes = len(np.unique(y_train))
print(f"Number of sentiment classes detected: {num_classes}")

# Convert integer labels to one-hot encoded vectors.
# This is a required step for multi-class classification in Keras.
y_train_encoded = to_categorical(y_train, num_classes=num_classes)
y_test_encoded = to_categorical(y_test, num_classes=num_classes)

# Standardize the features
# Neural networks are highly sensitive to feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build the Neural Network model for multi-class classification
print("\nBuilding Neural Network model...")
model = Sequential()
# Input layer
model.add(Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)))
model.add(Dropout(0.2))
# Hidden layer
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
# Output layer: The number of units must equal the number of classes.
# The activation is 'softmax' to output probabilities for each class.
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
# Using 'categorical_crossentropy' for multi-class classification.
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
print("Training Neural Network...")
history = model.fit(
    X_train_scaled,
    y_train_encoded, # Use the one-hot encoded labels for training
    epochs=10,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)

# Make predictions on the test set
# The model will output probabilities for each class
y_pred_probs = model.predict(X_test_scaled)
# We take the class with the highest probability
y_pred = np.argmax(y_pred_probs, axis=1)

# Print classification report to evaluate the model
# We use the original, non-encoded y_test for the classification report
print("\n--- Neural Network Classification Report ---")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.utils import class_weight
#from data_preprocessing import get_preprocessed_data

# Load the preprocessed data
X_train, X_test, y_train, y_test = get_preprocessed_data()

# --- FIX: Calculate and use class weights to handle imbalance ---
# Get the unique classes and their counts
class_weights_dict = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
# Convert the weights to a dictionary mapping class indices to weights
class_weights = {i: weight for i, weight in enumerate(class_weights_dict)}
print(f"\nClass weights for imbalance correction: {class_weights}")
# ---------------------------------------------------------------

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build the Neural Network model
print("\nBuilding Neural Network model...")
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
# Output layer for BINARY classification
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model with class weights
print("Training Neural Network with class weights...")
history = model.fit(
    X_train_scaled,
    y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.1,
    verbose=1,
    class_weight=class_weights # Pass the class weights here
)

# Make predictions
y_pred_probs = model.predict(X_test_scaled)
y_pred = (y_pred_probs > 0.5).astype("int32")

# Print classification report
# Note: The report will use the factorized labels (0 and 1)
print("\n--- Neural Network Classification Report (with Class Weights) ---")
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Embedding, LSTM, concatenate
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# --- 1. Load and Split Your Actual Data ---
# IMPORTANT: Replace this section with your actual data loading.
# This example assumes your data is already in a pandas DataFrame called `merged_game_data`.
# You should load it from your file (e.g., CSV) here.

# Placeholder for your data loading. Uncomment and modify as needed.
# merged_game_data = pd.read_csv('your_dataset_file.csv')

# --- Dummy Data for Demonstration (DELETE or comment out for real use) ---
# We'll use dummy data to make the script runnable for testing.
num_samples = 229774
num_numerical_features = 20
numerical_data = np.random.rand(num_samples, num_numerical_features).astype(np.float32)
positive_reviews = ["I love this game it's so much fun.", "Excellent game, highly recommend it.", "The graphics are amazing and the gameplay is smooth.", "Best game I have ever played, a must-buy."]
negative_reviews = ["This game is full of bugs and glitches.", "I hate this game, it's not fun at all.", "The worst experience, total waste of money.", "Terrible gameplay, do not recommend."]
labels = np.random.choice([0, 1], size=num_samples, p=[0.2, 0.8]).astype(np.int32)
text_data = np.array([np.random.choice(positive_reviews) if label == 1 else np.random.choice(negative_reviews) for label in labels])

# Create a dummy DataFrame to mimic your data structure
merged_game_data = pd.DataFrame(numerical_data, columns=[f'feature_{i}' for i in range(num_numerical_features)])
merged_game_data['review'] = text_data
merged_game_data['sentiment'] = labels
# --- End of Dummy Data Section ---

# Define the columns to use
numerical_cols = [col for col in merged_game_data.columns if col.startswith('feature_')]
text_col = 'review'
target_col = 'sentiment'

# Separate features (X) and target (y)
X_numerical = merged_game_data[numerical_cols].values
X_text = merged_game_data[text_col].values
y = merged_game_data[target_col].values

# Split the data into training and testing sets
numerical_train, numerical_test, text_train, text_test, y_train, y_test = train_test_split(
    X_numerical, X_text, y, test_size=0.2, random_state=42, stratify=y
)

# --- 2. Preprocess Numerical and Text Data ---
def preprocess_numerical(data_train, data_test):
    """Standardizes numerical data using StandardScaler."""
    scaler = StandardScaler()
    scaled_train = scaler.fit_transform(data_train)
    scaled_test = scaler.transform(data_test)
    return scaled_train, scaled_test

def preprocess_text(data_train, data_test):
    """Tokenizes and pads text data for the model."""
    max_words = 10000
    tokenizer = Tokenizer(num_words=max_words, oov_token="<unk>")
    tokenizer.fit_on_texts(data_train)

    train_sequences = tokenizer.texts_to_sequences(data_train)
    test_sequences = tokenizer.texts_to_sequences(data_test)

    max_sequence_length = 50
    padded_train = pad_sequences(train_sequences, maxlen=max_sequence_length, padding='post', truncating='post')
    padded_test = pad_sequences(test_sequences, maxlen=max_sequence_length, padding='post', truncating='post')

    return padded_train, padded_test, tokenizer

print("Preprocessing data...")
scaled_numerical_train, scaled_numerical_test = preprocess_numerical(numerical_train, numerical_test)
padded_text_train, padded_text_test, tokenizer = preprocess_text(text_train, text_test)

# --- 3. Build the Hybrid Model Architecture ---
def build_hybrid_model(numerical_input_shape, text_vocab_size, text_input_length):
    """
    Builds a multi-input neural network with separate branches for numerical and text data.

    Returns:
        tf.keras.Model: The compiled hybrid model.
    """
    # Branch 1: Numerical Input
    numerical_input = Input(shape=numerical_input_shape, name='numerical_input')
    numerical_dense = Dense(64, activation='relu')(numerical_input)

    # Branch 2: Text Input
    text_input = Input(shape=(text_input_length,), name='text_input')
    embedding_layer = Embedding(input_dim=text_vocab_size, output_dim=128)(text_input)
    # Using an LSTM layer to process the sequence
    text_lstm = LSTM(64)(embedding_layer)

    # Merge the two branches
    merged = concatenate([numerical_dense, text_lstm])

    # Final classification layers
    dense_1 = Dense(64, activation='relu')(merged)
    output_layer = Dense(1, activation='sigmoid')(dense_1)

    # Define the model with two inputs and one output
    model = Model(inputs=[numerical_input, text_input], outputs=output_layer)

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    return model

# Build the model using the preprocessed data shapes
vocab_size = len(tokenizer.word_index) + 1
hybrid_model = build_hybrid_model(
    numerical_input_shape=(scaled_numerical_train.shape[1],),
    text_vocab_size=vocab_size,
    text_input_length=padded_text_train.shape[1]
)

print("\nModel Architecture:")
print(hybrid_model.summary())

# --- 4. Train and Evaluate the Model ---
# Define class weights for imbalance correction (if your dataset is imbalanced)
# These values are based on an imbalanced dataset, but you should re-calculate
# them based on your actual data's class distribution.
class_weights = {0: 3.145, 1: 0.594}

print("\nTraining hybrid model...")
history = hybrid_model.fit(
    {'numerical_input': scaled_numerical_train, 'text_input': padded_text_train},
    y_train,
    epochs=10,
    batch_size=32,
    validation_data=({'numerical_input': scaled_numerical_test, 'text_input': padded_text_test}, y_test),
    class_weight=class_weights
)

print("\nEvaluating hybrid model...")
loss, accuracy = hybrid_model.evaluate(
    {'numerical_input': scaled_numerical_test, 'text_input': padded_text_test},
    y_test
)

print(f"\nFinal Test Loss: {loss:.4f}")
print(f"Final Test Accuracy: {accuracy:.4f}")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Embedding, LSTM, concatenate
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight

# --- 1. Load Your Actual Data and Sanity Check ---
print("Loading data...")
try:
    # IMPORTANT: Replace 'merged_game_data.csv' with the path to your actual file.
    merged_game_data = pd.read_csv('/content/drive/MyDrive/merged_game_data.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("ERROR: CSV file not found. Please check the file path.")
    import sys
    sys.exit()

# --- STEP 1: DEFINE YOUR COLUMNS ---
# Only include numerical columns in this list. The 'review' column is handled separately.
numerical_cols = ['voted_up', 'votes_up', 'votes_funny', 'author_playtime_forever',
                  'word_count', 'score_rank', 'positive', 'negative', 'average_forever', 'average_2weeks', 'median_forever',
                  'median_2weeks', 'initialprice', 'discount', 'ccu']
text_col = 'review'
target_col = 'sentiment'
# --- END STEP 1 ---

print("\n--- Data Sanity Check ---")
print(f"DataFrame shape: {merged_game_data.shape}")
print("All columns:", merged_game_data.columns.tolist())

# Check for correct column names
if not all(col in merged_game_data.columns for col in numerical_cols):
    print("ERROR: One or more numerical columns not found in DataFrame.")
    print("Please update the 'numerical_cols' list in the script.")
    import sys
    sys.exit()
if text_col not in merged_game_data.columns:
    print(f"ERROR: Text column '{text_col}' not found.")
    import sys
    sys.exit()
if target_col not in merged_game_data.columns:
    print(f"ERROR: Target column '{target_col}' not found.")
    import sys
    sys.exit()
if merged_game_data[target_col].nunique() < 2:
    print(f"ERROR: Target column '{target_col}' must have at least two unique classes.")
    import sys
    sys.exit()

# --- STEP 2: CONVERT STRING LABELS TO NUMBERS ---
# The model requires numerical labels (0 and 1) for training.
# We'll map 'positive' to 1 and 'neutral' to 0.
label_mapping = {'positive': 1, 'neutral': 0}
merged_game_data[target_col] = merged_game_data[target_col].map(label_mapping)
print("\nSentiment labels converted to numbers.")

# Check for missing values in critical columns
print(f"Missing values in '{text_col}': {merged_game_data[text_col].isnull().sum()}")
print(f"Missing values in '{target_col}': {merged_game_data[target_col].isnull().sum()}")

# Check the distribution of your new numerical target variable
print("\nTarget variable (sentiment) distribution:")
print(merged_game_data[target_col].value_counts())

# If the sanity checks pass, proceed with splitting and preprocessing
# Separate features (X) and target (y)
X_numerical = merged_game_data[numerical_cols].values
X_text = merged_game_data[text_col].values
y = merged_game_data[target_col].values

# Split the data first, before any preprocessing!
numerical_train, numerical_test, text_train, text_test, y_train, y_test = train_test_split(
    X_numerical, X_text, y, test_size=0.2, random_state=42, stratify=y
)
print("\nData split into training and testing sets successfully.")


# --- 2. Preprocess Numerical and Text Data ---
def preprocess_numerical(data_train, data_test):
    """
    Handles missing values and standardizes numerical data.
    """
    # Replace NaN values with the median of the training data
    median_train = np.nanmedian(data_train, axis=0)
    data_train = np.nan_to_num(data_train, nan=median_train)
    data_test = np.nan_to_num(data_test, nan=median_train)

    scaler = StandardScaler()
    scaled_train = scaler.fit_transform(data_train)
    scaled_test = scaler.transform(data_test)
    return scaled_train, scaled_test

def preprocess_text(data_train, data_test):
    """Tokenizes and pads text data for the model."""
    max_words = 10000
    tokenizer = Tokenizer(num_words=max_words, oov_token="<unk>")
    tokenizer.fit_on_texts(data_train)

    train_sequences = tokenizer.texts_to_sequences(data_train)
    test_sequences = tokenizer.texts_to_sequences(data_test)

    max_sequence_length = 50
    padded_train = pad_sequences(train_sequences, maxlen=max_sequence_length, padding='post', truncating='post')
    padded_test = pad_sequences(test_sequences, maxlen=max_sequence_length, padding='post', truncating='post')

    return padded_train, padded_test, tokenizer

print("\nPreprocessing numerical data...")
scaled_numerical_train, scaled_numerical_test = preprocess_numerical(numerical_train, numerical_test)

print("Preprocessing text data...")
padded_text_train, padded_text_test, tokenizer = preprocess_text(text_train, text_test)


# --- 3. Build the Hybrid Model Architecture ---
def build_hybrid_model(numerical_input_shape, text_vocab_size, text_input_length):
    """
    Builds a multi-input neural network with separate branches for numerical and text data.

    Returns:
        tf.keras.Model: The compiled hybrid model.
    """
    numerical_input = Input(shape=numerical_input_shape, name='numerical_input')
    numerical_dense = Dense(64, activation='relu')(numerical_input)

    text_input = Input(shape=(text_input_length,), name='text_input')
    embedding_layer = Embedding(input_dim=text_vocab_size, output_dim=128)(text_input)
    text_lstm = LSTM(64)(embedding_layer)

    merged = concatenate([numerical_dense, text_lstm])

    dense_1 = Dense(64, activation='relu')(merged)
    output_layer = Dense(1, activation='sigmoid')(dense_1)

    # Use a learning rate and add gradient clipping for stability
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001, clipnorm=1.0)

    model = Model(inputs=[numerical_input, text_input], outputs=output_layer)

    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

    return model

vocab_size = len(tokenizer.word_index) + 1
hybrid_model = build_hybrid_model(
    numerical_input_shape=(scaled_numerical_train.shape[1],),
    text_vocab_size=vocab_size,
    text_input_length=padded_text_train.shape[1]
)

print("\nModel Architecture:")
print(hybrid_model.summary())

# --- 4. Train and Evaluate the Model ---
# Dynamically calculate class weights from the training data to handle imbalance
unique_classes = np.unique(y_train)
class_weights_array = compute_class_weight('balanced', classes=unique_classes, y=y_train)
class_weights = dict(zip(unique_classes, class_weights_array))

print("\nCalculated Class Weights:", class_weights)
print("Training hybrid model...")
history = hybrid_model.fit(
    {'numerical_input': scaled_numerical_train, 'text_input': padded_text_train},
    y_train,
    epochs=10,
    batch_size=32,
    validation_data=({'numerical_input': scaled_numerical_test, 'text_input': padded_text_test}, y_test),
    class_weight=class_weights
)

print("\nEvaluating hybrid model...")
loss, accuracy = hybrid_model.evaluate(
    {'numerical_input': scaled_numerical_test, 'text_input': padded_text_test},
    y_test
)

print(f"\nFinal Test Loss: {loss:.4f}")
print(f"Final Test Accuracy: {accuracy:.4f}")
```
