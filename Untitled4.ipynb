```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive

# --- 1. Get the dataset from Google Drive ---
# Mount Google Drive to access files
print("Mounting Google Drive...")
drive.mount('/content/drive')

# UPDATE THIS PATH: Change 'my_dataset.csv' to the actual path/filename of your dataset.
FILE_PATH = '/content/drive/MyDrive/steam_sentiment/cleaned_all_reviews.csv'
PRICE_COLUMN = 'price' # Update this if your price column has a different name

try:
    # Load the dataset
    df = pd.read_csv(FILE_PATH)
    print(f"Dataset loaded successfully from: {FILE_PATH}")
    print(f"Total number of records: {len(df)}")

    # Ensure the price column exists and is numeric
    if PRICE_COLUMN not in df.columns:
        print(f"\nERROR: Column '{PRICE_COLUMN}' not found in the dataset. Please update PRICE_COLUMN.")
    else:
        # Convert price column to numeric, errors='coerce' turns non-numeric values into NaN
        df[PRICE_COLUMN] = pd.to_numeric(df[PRICE_COLUMN], errors='coerce')
        # Drop rows where price is missing or not a valid number
        df.dropna(subset=[PRICE_COLUMN], inplace=True)

        # --- 2. Count the number of reviews per price group ---
        # Define the bins and labels for the price groups:
        # (0-20 low, 20-40 mid, 40-max high)
        # Note: Bins are exclusive on the left and inclusive on the right by default: (low, high]
        bins = [0, 20, 40, df[PRICE_COLUMN].max() + 1] # +1 ensures the max value is included
        labels = ['1-20 (Low)', '20-40 (Mid)', '40-Max (High)']

        # Create the new categorical column
        df['price_group'] = pd.cut(
            df[PRICE_COLUMN],
            bins=bins,
            labels=labels,
            right=True, # Make intervals inclusive on the right (e.g., price=20 is 'Low')
            include_lowest=True # Include the lowest price (0)
        )

        # Count the number of reviews in each group
        review_counts = df['price_group'].value_counts().reindex(labels, fill_value=0)

        print("\nReview Counts by Price Group:")
        print(review_counts.to_string())

        # --- 3. Show it on a bar chart ---
        plt.figure(figsize=(8, 5))
        review_counts.plot(kind='bar', color=['#4CAF50', '#FFC107', '#F44336'])
        plt.title('Number of Reviews per Price Group', fontsize=14)
        plt.xlabel('Price Group', fontsize=12)
        plt.ylabel('Number of Reviews', fontsize=12)
        plt.xticks(rotation=0) # Keep labels horizontal
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

except FileNotFoundError:
    print(f"\nERROR: File not found at {FILE_PATH}. Please check the path and file name.")
except Exception as e:
    print(f"\nAn error occurred: {e}")
```

```python
# imports
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

# --- 1) clean price column (works if price is numeric or strings like '$1.99', 'Free', '0', '1,299.00')
def clean_price(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float, np.number)):
        return float(x)
    s = str(x).strip()
    if s.lower() in ['free', '0', '0.0', '0.00', 'free.']:
        return 0.0
    # remove any non-digit/period characters (currency symbols, commas, whitespace)
    s2 = re.sub(r'[^\d\.]', '', s)
    try:
        return float(s2) if s2 != '' else np.nan
    except:
        return np.nan

df['price_clean'] = df['price'].apply(clean_price)

# --- 2) create price segments
labels = ['1-20 (Low)', '20-40 (Mid)', '40-max (High)']
bins = [0, 20, 40, np.inf]   # last bin is 40 to infinity
df['price_segment'] = pd.cut(df['price_clean'], bins=bins, labels=labels,
                             right=False, include_lowest=True)

# --- 3) detect non-empty reviews
has_review = df['review'].notna() & df['review'].astype(str).str.strip().ne('')

# --- 4) count reviews per segment
review_counts = (df.loc[has_review]
                 .groupby('price_segment')
                 .size()
                 .reindex(labels, fill_value=0)
                 .reset_index(name='review_count'))

# --- optional: count unique apps per segment
unique_app_counts = (df.groupby('price_segment')['appid']
                      .nunique()
                      .reindex(labels, fill_value=0)
                      .reset_index(name='unique_app_count'))

# --- merge results and show
summary = review_counts.merge(unique_app_counts, on='price_segment')
print(summary)

# --- optional: quick bar plot of review counts
plt.figure(figsize=(7,4))
plt.bar(summary['price_segment'].astype(str), summary['review_count'])
plt.title('Number of reviews per price segment')
plt.ylabel('Review count')
plt.xlabel('Price segment')
plt.xticks(rotation=10)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---- Parameters you can tune
min_samples = 5   # only show price points with at least this many reviews (change to 1 to include all)
top_k = 3         # how many top sentiment points to highlight

# ---- Make a working copy so we don't mutate your original df unexpectedly
_df = df.copy()

# ---- Ensure price_clean & sentiment exist and are numeric
for col in ['price_clean', 'clean_review_sentiment_num', 'appid']:
    if col not in _df.columns:
        raise KeyError(f"Missing required column: {col}")

_df['price_clean'] = pd.to_numeric(_df['price_clean'], errors='coerce')
_df['clean_review_sentiment_num'] = pd.to_numeric(_df['clean_review_sentiment_num'], errors='coerce')

# ---- Build price_point (floor to integer dollar)
_df = _df.dropna(subset=['price_clean', 'clean_review_sentiment_num', 'appid'])
_df['price_point'] = np.floor(_df['price_clean']).astype(int)

# ---- Aggregate: mean sentiment & review count per price_point
pp_summary = (_df
              .groupby('price_point')['clean_review_sentiment_num']
              .agg(mean_sentiment='mean', review_count='count')
              .reset_index()
              .sort_values('price_point'))

# ---- Aggregate: unique games per price_point (based on appid)
games_per_price = (_df
                   .groupby('price_point')['appid']
                   .nunique()
                   .reset_index(name='unique_game_count')
                   .sort_values('price_point'))

# ---- Merge the tables so we have mean_sentiment, review_count, unique_game_count in one frame
merged = pp_summary.merge(games_per_price, on='price_point', how='left').fillna(0)

# ---- Filter low-sample price points (so super-rare prices don't dominate)
merged_plot = merged[merged['review_count'] >= min_samples].copy()
if merged_plot.empty:
    raise ValueError(f"No price points meet min_samples={min_samples}. Lower min_samples or check data.")

# ---- Find top-k price points by mean sentiment among the plotted points
top_points = merged_plot.nlargest(top_k, 'mean_sentiment')['price_point'].values

# ---- Plotting: mean sentiment (line), highlight top-k points, and unique_game_count as bars on secondary axis
fig, ax1 = plt.subplots(figsize=(12,6))

# line for mean sentiment
ax1.plot(merged_plot['price_point'], merged_plot['mean_sentiment'],
         marker='o', linewidth=1.5, label='Avg sentiment (mean)', zorder=2)

# plot all points as small markers (base)
ax1.scatter(merged_plot['price_point'], merged_plot['mean_sentiment'],
            s=40, alpha=0.9, zorder=3)

# highlight top-k points with different marker & larger size
top_df = merged_plot[merged_plot['price_point'].isin(top_points)]
ax1.scatter(top_df['price_point'], top_df['mean_sentiment'],
            s=160, marker='D', edgecolor='black', linewidth=0.8,
            label=f'Top {top_k} sentiment points', zorder=4)

ax1.set_xlabel('Price point (Dollars)')
ax1.set_ylabel('Average numeric sentiment')
ax1.set_title('Average sentiment by integer price point')

# secondary axis: number of unique games (bars)
ax2 = ax1.twinx()
bar_width = 0.6
ax2.bar(merged_plot['price_point'] - bar_width*0.0,  # no offset, overlayed but on second axis
        merged_plot['unique_game_count'],
        width=bar_width, alpha=0.25, label='Unique games', zorder=1)

ax2.set_ylabel('Number of unique games')

# tidy x ticks (show all plotted integer price points)
ax1.set_xticks(merged_plot['price_point'])
ax1.set_xticklabels(merged_plot['price_point'].astype(str), rotation=45)

# construct combined legend
lines_labels = [l.get_label() for l in ax1.get_legend_handles_labels()[0:2]]
hlines = ax1.get_legend_handles_labels()[0]
hbars = ax2.get_legend_handles_labels()[0]
# Use a simple legend that combines handles from both axes
handles = ax1.get_legend_handles_labels()[0] + ax2.get_legend_handles_labels()[0]
labels = ax1.get_legend_handles_labels()[1] + ax2.get_legend_handles_labels()[1]
ax1.legend(handles, labels, loc='upper left')

plt.tight_layout()
plt.show()

# ---- Print a small summary table of the plotted points (optional)
display_cols = ['price_point', 'mean_sentiment', 'review_count', 'unique_game_count']
print("Top plotted price points by mean sentiment (and their stats):")
print(merged_plot[display_cols].sort_values('mean_sentiment', ascending=False).head(10).to_string(index=False))
```
