```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/cleaned_all_reviews.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Define negative reviews (1â€“2 stars)
negative_df = df[df["clean_review_sentiment_num"] <= 2]

# Total reviews per price segment
total_counts = df.groupby("price_section").size()

# Negative reviews per price segment
negative_counts = negative_df.groupby("price_section").size()

# Percentage calculation
negative_percentage = (negative_counts / total_counts * 100).fillna(0)

# Summary table
summary = pd.DataFrame({
    "Total Reviews": total_counts,
    "Negative Reviews": negative_counts,
    "Negative Percentage (%)": negative_percentage.round(2)
})

summary
```

```python
negative_percentage.plot(kind="bar")
plt.xlabel("Price Section")
plt.ylabel("Negative Review Percentage (%)")
plt.title("Negative Review Percentage by Price Segment")
plt.show()
```
